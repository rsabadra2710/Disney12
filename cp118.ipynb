```python
import pandas as pd
import plotly.express as px
e=pd.read_csv('/content/stars.csv')
r=px.scatter(e, x="Size",y="Light")
r.show()
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

w=e.iloc[:,[0,1]].values
q=[]

for i in range(1,10):
  k=KMeans(i)
  k.fit(w)
  z=k.inertia_
  q.append(z)

x=range(1,10)
plt.plot(x,q)
plt.title('elbow')
plt.xlabel('number of clusters')
plt.ylabel('inertia')
```

```python
kmeans = KMeans(n_clusters = 3)
y_kmeans = kmeans.fit_predict(w) 
sns.scatterplot(w[y_kmeans == 0, 0], w[y_kmeans == 0, 1], color = 'yellow', label = 'Cluster 1')
sns.scatterplot(w[y_kmeans == 1, 0], w[y_kmeans == 1, 1], color = 'blue', label = 'Cluster 2') 
sns.scatterplot(w[y_kmeans == 2, 0], w[y_kmeans == 2, 1], color = 'green', label = 'Cluster 3') 
sns.scatterplot(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'red', label = 'Centroids')
plt.title('Clusters of Flowers') 
plt.xlabel('Size')
plt.ylabel('Light') 
plt.show()
```
