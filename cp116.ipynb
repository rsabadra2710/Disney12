```python
import pandas as pd
import plotly.express as px
e=pd.read_csv('/content/Admission_Predict.csv')
salary=e["GRE Score"].tolist()
purchased=e["University Rating"].tolist()
x=px.scatter(x=salary, y=purchased)
x.show()
```

```python
import plotly.graph_objects as pgo
salary=e["GRE Score"].tolist()
age=e["CGPA"].tolist()
purchased=e["University Rating"].tolist()
colors=[]
for i in purchased:
  if i==1:
    colors.append("green")
  else:
   colors.append("red")

d=pgo.Figure(data=pgo.Scatter(
    x=salary, y=age, mode='markers', marker=dict(color=colors)
))
d.show()
```
