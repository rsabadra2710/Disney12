```python
import pandas as pd
import plotly.express as px
r=pd.read_csv('/content/titanic.csv')
e=px.scatter(r, x='PassengerId', y='Sex')
e.show()
```

```python
features=['PassengerId','Pclass','Sex','Age','SibSp','Parch']
x=r[features]
y=r['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
z=DecisionTreeClassifier()

z=z.fit(x_train,y_train)
p=z.predict(x_test)
print('accuracy', metrics.accuracy_score(y_test,p))
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from six import StringIO
from IPython.display import Image

q=StringIO()
export_graphviz(z,out_file=q,feature_names=features,class_names=['0','1'], filled=True,rounded=True)
graph=pydotplus.graph_from_dot_data(q.getvalue())
Image(graph.create_png())
```
